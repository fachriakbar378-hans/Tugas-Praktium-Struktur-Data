```python
class Node:
  def __init__(self, dataval = None):
    self.dataval = dataval
    self.nextval = None

# Create Node
n1 = Node("Januari")
n2 = Node("Februari")

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the Linked List
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

# Funtion to add node in the middle
  def AddInBetween(self, mid_node, newdata):
    if mid_node is None:
      print("The mentioned node is absent")
      return

    NewNode = Node(newdata)
    NewNode.nextval = mid_node.nextval
    mid_node.nextval = NewNode

Li = LinkedList()
Li.headval = n1

#Link first Node to second node
Li.headval.nextval = n2

# Add Node after n1
Li.AddInBetween(n1,"Middle")
Li.listprint()
```

Penyisipan di Akhir

```python
class Node:
  def __init__(self, dataval = None):
    self.dataval = dataval
    self.nextval = None

# Create Node
n1 = Node("Januari")
n2 = Node("Februari")

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the Linked List
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

# Funtion to add node in the End
  def AddInEnd(self, newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return

    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval = NewNode

Li = LinkedList()
Li.headval = n1

#Link first Node to second node
Li.headval.nextval = n2

# Add Node in the End
Li.AddInEnd("The Last")
Li.listprint()
```

Menghapus Node dengan Kata Kunci

```python
class Node:
  def __init__(self, dataval = None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the Linked List
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

# Function to add Node in the begining
  def AddBegining(self, newdata):
    NewNode = Node(newdata)

# Update the new nodes ext val tox eisting node
    NewNode.nextval = self.headval
    self.headval = NewNode

# Function to remove node by key
  def RemoveNode(self, Removekey):
    HeadVal = self.headval
    if (HeadVal is not None):
      if (HeadVal.dataval == Removekey):
        self.headval = HeadVal.next
        HeadVal = None
        return

    while(HeadVal is not None):
      if HeadVal.dataval == Removekey:
        break
      prev = HeadVal
      HeadVal = HeadVal.nextval
    if (HeadVal == None):
      return
    prev.nextval = HeadVal.nextval
    HeadVal = None

# Create linked list named Li
Li = LinkedList()

# Add Node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Februari")
print("Sebelum dilakukan Remove")
Li.listprint()
print("==========================")

# Remove by key = "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan Remove dengan kunci 'maret'")
Li.listprint()
```

Menghapus Node Terakhir

```python
class Node:
  def __init__(self, dataval = None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the Linked List
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

# Function to add Node in the begining
  def AddBegining(self, newdata):
    NewNode = Node(newdata)

# Update the new nodes ext val tox eisting node
    NewNode.nextval = self.headval
    self.headval = NewNode

# Function to remove last node
  def RemoveEnd(self):
    last = self.headval
    while (last is not None):
      if last.nextval == None:
        break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None

# Create linked list named Li
Li = LinkedList()

# Add Node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Februari")
print("Sebelum dilakukan Remove")
Li.listprint()
print("==========================")

# Remove Last Node
Li.RemoveEnd()
print("Setelah dilakukan Remove node terakhir")
Li.listprint()
```

```python
# 1. Buatlah double linked list dengan 3 node didalamnya, kemudian lakukan
# penyisipan tengah dan hapus dengan kata kunci !

# Double linked list dengan 3 node
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


# Fungsi untuk menampilkan linked list
def print_list(head):
    current = head
    while current is not None:
        print(current.data, end=" <-> ")
        current = current.next
    print("None")


# Membuat node awal
n1 = Node("Firefly")
n2 = Node("Hans")
n3 = Node("Castorice")

n1.next = n2
n2.prev = n1

n2.next = n3
n3.prev = n2

head = n1

print("Linked List Awal:")
print_list(head)


# Menyisipkan ke tengah
newNode = Node("Evernight")
newNode.prev = n1
newNode.next = n2

n1.next = newNode
n2.prev = newNode

print("\nSetelah Penyisipan di Tengah:")
print_list(head)


# Menghapus node dengan kata kunci
def delete_by_key(head, key):

    current = head

    while current is not None:
        if current.data == key:

            if current == head:
                head = current.next
                if head is not None:
                    head.prev = None

            elif current.next is None:
                current.prev.next = None

            else:
                current.prev.next = current.next
                current.next.prev = current.prev

            current.prev = None
            current.next = None
            break

        current = current.next

    return head


# Hapus node dengan kata kunci
head = delete_by_key(head, "Evernight")

print("\nSetelah Menghapus Node dengan Kata Kunci 'Evernight':")
print_list(head)
```

```python
# 2. Buatlah circular linked list dengan 3 node didalamnya,
# kemudian lakukan penyisipan akhir dan hapus akhir !

# Struktur node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Fungsi untuk menampilkan circular linked list
def print_circular_list(head):
    if head is None:
        print("List kosong")
        return

    current = head
    while True:
        print(current.data, end=" -> ")
        current = current.next
        if current == head:
            break
    print("(back to head)")

# 1. Circular Linked List Awal (3 Node)
n1 = Node("Windah")
n2 = Node("Regi")
n3 = Node("Reza")

n1.next = n2
n2.next = n3
n3.next = n1   # circular

head = n1

print("Circular Linked List Awal:")
print_circular_list(head)

# 2. Penyisipan di Akhir
newNode = Node("Auditore")

current = head
while current.next != head:
    current = current.next

current.next = newNode
newNode.next = head

print("\nSetelah Penyisipan di Akhir:")
print_circular_list(head)

# 3. Hapus Node Akhir
def delete_end_circular_list(head):
    if head is None:
        return None

    # Jika hanya satu node
    if head.next == head:
        return None

    current = head
    while current.next.next != head:
        current = current.next

    # current adalah node sebelum terakhir
    current.next = head
    return head


head = delete_end_circular_list(head)

print("\nSetelah Hapus Akhir:")
print_circular_list(head)
```
