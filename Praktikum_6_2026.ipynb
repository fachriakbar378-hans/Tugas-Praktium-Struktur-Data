Insertion Sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = bubble_sort(data.copy()) # Menggunakan copy agar data asli tidak berubah
print("Data sebelum sorting:", data)
print("Data setelah Bubble Sort:", sorted_data)
```

Selection Sort

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = selection_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Selection Sort:", sorted_data)
```

Insertion Sort

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Insertion Sort:", sorted_data)
```

Merge Sort

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]
        merge_sort(left_half)
        merge_sort(right_half)
        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr
# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Merge Sort:", sorted_data)
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Quick Sort:", sorted_data)
```

```python
# Data pelanggan
pelanggan = [
    {"nama": "Firefly", "umur": 20, "total_belanja": 150000},
    {"nama": "Kafka", "umur": 35, "total_belanja": 300000},
    {"nama": "Hans", "umur": 21, "total_belanja": 200000},
    {"nama": "Castorice", "umur": 21, "total_belanja": 250000}
]

# 1. Sort berdasarkan nama (A-Z)
sort_nama = sorted(pelanggan, key=lambda x: x["nama"])

# 2. Sort berdasarkan umur (kecil ke besar)
sort_umur = sorted(pelanggan, key=lambda x: x["umur"])

# 3. Sort berdasarkan total belanja (besar ke kecil)
sort_belanja = sorted(pelanggan, key=lambda x: x["total_belanja"], reverse=True)

# Output
print("=== Sort Nama (A-Z) ===")
for p in sort_nama:
    print(p)

print("\n=== Sort Umur ===")
for p in sort_umur:
    print(p)

print("\n=== Sort Total Belanja (Terbesar) ===")
for p in sort_belanja:
    print(p)
```
