Tugas 1: Program Struktur Binary Tree Manual

```python
# Struktur Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# =========================
# Data identitas
# =========================

nama = "Hans Kusuma Agung"
npm = "2505060077"      # contoh NPM
tanggal_lahir = "02-08-2006"

# Menghitung jumlah huruf nama (tanpa spasi)
jumlah_huruf = len(nama.replace(" ", ""))

# Menghitung jumlah huruf vokal
vokal = "aiueoAIUEO"
jumlah_vokal = sum(1 for huruf in nama if huruf in vokal)

# Dua digit terakhir NPM
dua_digit_npm = int(npm[-2:])

# Ambil tanggal dan bulan lahir
tanggal = int(tanggal_lahir.split("-")[0])
bulan = int(tanggal_lahir.split("-")[1])


# =========================
# Membuat Binary Tree
# =========================

root = Node(jumlah_huruf)

# Anak kiri dan kanan root
root.left = Node(jumlah_vokal)
root.right = Node(dua_digit_npm)

# Level berikutnya
root.left.left = Node(tanggal)
root.right.right = Node(bulan)


# =========================
# Inorder Traversal
# =========================

def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)


# =========================
# Output
# =========================

print("Hasil Inorder Traversal:")
inorder(root)
```

Tugas 2: Program Binary Search Tree (BST)

```python
# Program Binary Search Tree (BST)

# =========================================
# Membuat class Node
# =========================================
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# =========================================
# Fungsi untuk menambahkan data ke BST
# =========================================
def insert(root, data):
    if root is None:
        return Node(data)

    # Jika data lebih kecil, masuk ke kiri
    if data < root.data:
        root.left = insert(root.left, data)

    # Jika data lebih besar, masuk ke kanan
    elif data > root.data:
        root.right = insert(root.right, data)

    return root


# =========================================
# Traversal Inorder
# Left -> Root -> Right
# =========================================
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)


# =========================================
# Traversal Preorder
# Root -> Left -> Right
# =========================================
def preorder(root):
    if root:
        print(root.data, end=" ")
        preorder(root.left)
        preorder(root.right)


# =========================================
# Traversal Postorder
# Left -> Right -> Root
# =========================================
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data, end=" ")


# =========================================
# Fungsi pencarian data pada BST
# =========================================
def search(root, key):
    if root is None:
        return False

    if root.data == key:
        return True

    elif key < root.data:
        return search(root.left, key)

    else:
        return search(root.right, key)


# =========================================
# Data unik berdasarkan identitas
# =========================================

# Contoh:
# Nama depan = Hans
# ASCII H = 72
# ASCII a = 97
# NPM = 2505060077

data = [
    277,   # tiga digit terakhir NPM
    2,     # tanggal lahir
    8,     # bulan lahir
    72,    # ASCII huruf pertama nama depan (H)
    97,    # ASCII huruf kedua nama depan (a)
    17,    # angka favorit
    45     # angka bebas
]


# =========================================
# Membentuk BST
# =========================================
root = None

for angka in data:
    root = insert(root, angka)


# =========================================
# Menampilkan traversal
# =========================================
print("Inorder Traversal:")
inorder(root)

print("\n\nPreorder Traversal:")
preorder(root)

print("\n\nPostorder Traversal:")
postorder(root)


# =========================================
# Pencarian data
# =========================================

cari1 = 77     # dua digit terakhir NPM
cari2 = 100    # angka yang tidak ada dalam BST

print("\n\nHasil Pencarian:")

if search(root, cari1):
    print(cari1, "ditemukan dalam BST")
else:
    print(cari1, "tidak ditemukan dalam BST")

if search(root, cari2):
    print(cari2, "ditemukan dalam BST")
else:
    print(cari2, "tidak ditemukan dalam BST")
```
